```python
class estimator:

    def __init__(self, meas_noise = 222.40084787, proc_noise = 178.13354365):
        # set up convoyor belt
        belt_length = 62 #belt length in inches
        volume_length = 2 #volume length in inches
        
      
        # set up camera
        camera_left_loc = 30 #distance in inches from the beginning of the belt of pixel column 0
        camera_right_loc = 0 #distance in inches from the beginning of the belt of highest valued pixel column (i.e. right side)
        
        
        self.num_materials = 2
        self.num_volumes = int(belt_length/volume_length)

        self.horizon = 20
        
        self.im_width = 1280
        self.sigma_process = proc_noise
        self.sigma_measurement = meas_noise
        self.num_speeds = 8
        self.speed_control_ratio = 0.1
        self.speed_init = 1.1
        self.speed_request = self.speed_init
        self.speed_request_block = self.speed_init
        self.U_traj = self.speed_request * np.ones((self.horizon))

        
        self.state_dim = self.num_materials * self.num_volumes + 1
        self.state_measurement = np.zeros((self.state_dim))
        self.state_prediction = np.zeros((self.state_dim))
        self.state_estimate = np.zeros((self.state_dim))
        self.innovation = np.zeros((self.state_dim))

        self.state_estimate[-1] = self.speed_init
        self.state_hist = []
        # self.cov_hist
        #set up camera
        self.cam_left_vol_id = int(camera_left_loc/volume_length)
        self.cam_right_vol_id = int(camera_right_loc/volume_length)
        
        
        if self.cam_left_vol_id > self.cam_right_vol_id:
            self.dir = -1
        else:
            self.dir = 1

        
        self.cam_span = self.dir * (self.cam_right_vol_id - self.cam_left_vol_id)
        # generate measurement jacobians        
        self.cam_dim = self.num_materials * self.dir * (self.cam_right_vol_id - self.cam_left_vol_id)
        self.H_k = np.zeros((self.cam_dim, self.state_dim))

        for cam_loc in range(self.cam_span):
            for mat in range(self.num_materials):
    
                vol_id = (self.cam_left_vol_id-1)*self.num_materials + self.dir * (cam_loc * self.num_materials) + mat
                self.H_k[cam_loc*self.num_materials + mat, vol_id] = 1

        self.B = np.zeros((self.state_dim))
        self.B[-1] = self.speed_control_ratio
        
        self.A = self.generate_a_matrix()
        self.A_lin = self.A[0]
        self.A_sorted = np.zeros_like(self.A_lin)
        self.P = np.eye(self.state_dim) * self.sigma_process
        self.P_pred = np.eye(self.state_dim) * self.sigma_measurement
        self.R = np.eye(self.cam_dim) * self.sigma_measurement
        self.Q = np.eye(self.state_dim) * self.sigma_process
        
        self.K = np.zeros((self.state_dim, self.state_dim))

        #set up pick zone
        self.sort_vec = np.zeros((self.state_dim))

        self.picks = []
        self.cum_sorts = 0
        

        item_0_start_id = 10
        item_0_end_id = 15
        self.update_rate = 12
        self.picks_per_sec = 1.5
        pick_width_0 = item_0_end_id - item_0_start_id
        picks_0 = []
        for i in range(pick_width_0):
            picks_0.append(item_0_start_id + i)
        self.picks.append(picks_0)

        self.delta = 0.05

        self.prices = [0.306, 0.01875] #units here are cents/unit.  PET: 12.25 c/lb, 3-7 bale 0.75 c/lb 40 pc/lb
        # self.picks_per_step_per_volume_max = 1.2/((item_0_end_id - item_0_start_id)*12) #one item per second, filter running at nominal ten hz
        
        self.iters = 10
        self.learning_rate = 0.05
        
    def unpack_image(self, msg):
        image = np.zeros((msg.height, msg.width, 3), dtype=np.uint8)        
        
        for row in range(msg.height):
            for i in range(msg.width):
                image[row, i, 2] = int(msg.data[msg.step*row + 3*i])
                image[row, i, 1] = int(msg.data[msg.step*row + 3*i + 1])
                image[row, i, 0] = int(msg.data[msg.step*row + 3*i + 2])
        return image
    
    def measurement_from_detections(self, msg):
        self.state_measurement = np.zeros((self.cam_dim))
        for obj_id in range(len(msg.bounding_boxes)):
            # first left-right flip
            start_vol = int(msg.bounding_boxes[obj_id].xmin/(self.im_width/self.cam_span))
            end_vol = int(msg.bounding_boxes[obj_id].xmax/((self.im_width+1)/self.cam_span))
            span = 1 + end_vol - start_vol
            # print('class: ', msg.bounding_boxes[obj_id].Class)
            # print('start_vol: ', start_vol, ', xmin: ', msg.bounding_boxes[obj_id].xmin)
            # # print()
            # # print()
            # print('end_vol: ', end_vol, ', xmax: ', msg.bounding_boxes[obj_id].xmax)
            # print('span: ', span)

            inds = np.zeros((span), dtype = int)
            for ind in range(span):
                inds[ind] = int((start_vol+ind)*self.num_materials)
            
                # if msg.bounding_boxes[obj_id].Class == 'Bottles':
                #     inds[ind] += 1
                if msg.bounding_boxes[obj_id].Class == 'Cans':
                    inds[ind] += 1
            # print('inds: ', inds)
            # print('state: ', self.state)
            self.state_measurement[inds] += 1/span
            # print('state: ', self.state)
            # print('sum of state: ', np.sum(self.state), '\n\n')
        # print(self.state_measurement)
        return self.state_measurement

    def get_state_prediction(self):        
        self.linearize_A()
        return self.A_lin @ self.state_estimate + self.B * self.speed_request

    def generate_a_matrix(self):
        A = np.zeros((self.state_dim, self.state_dim, self.num_speeds))
        for speed in range(self.num_speeds):
            for i in range((self.num_volumes-speed)*self.num_materials):
                A[i+(speed * self.num_materials), i,  speed] = 1
            for j in range(speed):
                for i in range(self.num_materials):
                    A[j*self.num_materials + i, i, speed] = 1
            A[-1, -1, speed] = 1-self.speed_control_ratio
        return A


    def cov_prediction(self):
        return self.A_lin @ self.P @ self.A_lin.T + self.Q

    def correction(self):
        self.state_hist.append(self.state_estimate)
        self.innovation = self.state_measurement - self.H_k @ self.state_prediction
        self.S_inv = np.linalg.inv(self.H_k @ self.P_pred @ self.H_k.T + self.R)
        self.K = self.P_pred @ self.H_k.T @ self.S_inv
        self.state_estimate = self.state_prediction + self.K @ self.innovation
        # print((self.K))
        self.P = (np.eye(self.state_dim) - self.K @ self.H_k) @ self.P_pred
        # plt.imshow(self.K, cmap='gray')
        # plt.show()
        # plt.imshow(self.innovation, cmap='gray')
        
        # plt.show()
        # plt.imshow(self.P, cmap='gray')

    def increment_filter(self):
        self.state_prediction = self.get_state_prediction()
        self.P_pred = self.cov_prediction()

        
        # self.state_from_detections(msg)
        self.correction()

    def linearize_A(self):
        floor = np.floor(self.state_estimate[-1])
        sort_vec = np.zeros((E.state_dim))
        # print(floor)
        ceil = floor + 1
        ratio = self.state_estimate[-1] - floor
        self.A_lin = self.A[:, :, int(floor)] * (1-ratio) + self.A[:, :, int(ceil)] * ratio
        self.get_sort_vec()
        self.A_lin = self.A_lin * (np.ones((self.state_dim)) - self.sort_vec)
        # plt.imshow(self.A_lin)
        # plt.show()
    
    def rollout_linearize_A(self, x):
        sort_vec = np.zeros((self.state_dim))
        floor = np.floor(x[-1])
        # print(floor)
        ceil = floor + 1
        ratio = x[-1] - floor
        A = self.A[:, :, int(floor)] * (1-ratio) + self.A[:, :, int(ceil)] * ratio
        
        for mat in range(len(self.picks)):
            for idx, i in enumerate(self.picks[mat]):
                sort_vec[i*self.num_materials + mat] = 1
            if sort_vec.T @ x > self.picks_per_sec / self.update_rate:
                sort_vec *= (self.picks_per_sec/self.update_rate)/(sort_vec.T @ x)
    
        A = A * (np.ones((self.state_dim)) - sort_vec)
        
        return A, sort_vec

    def rollout(self, x, U_traj):
        X_traj = np.zeros((self.horizon, self.state_dim))
        score = 0
        for step in range(self.horizon):
            A_rollout, sort_vec = self.rollout_linearize_A(x)
            if step == 0:
                X_traj[step] = A_rollout @ x + self.B * U_traj[step]
            else:
                X_traj[step] = A_rollout @ X_traj[step - 1] + self.B * U_traj[step]
            
            score_vec = self.prices[0] * sort_vec
            score_vec[(self.num_volumes-1)*self.num_materials : self.num_volumes*self.num_materials] = self.prices[1]

            score += score_vec @ X_traj[step]
        
        return score, X_traj
        
    def solve_opt_traj(self, x_init):

        #roll out nom traj, nom traj + delta
        U_traj_nom = self.U_traj
        grad = np.zeros_like(U_traj_nom)
        deltas = self.delta * np.diag(np.ones_like(U_traj_nom))
        lr = self.learning_rate
        score_logging = np.zeros((self.iters))
        U_traj_logging = np.zeros((self.iters, self.horizon))

        base_score, x_t = self.rollout(x_init, U_traj_nom)
        
        for i in range(self.iters):
            #get gradient about u_traj_nom
            
            for dir in range(self.horizon):
                s, x_t = self.rollout(x_init, U_traj_nom + deltas[dir])
                grad[dir] = (s-base_score)/self.delta
            U_traj_nom += lr * grad
            base_score, x_t = self.rollout(x_init, U_traj_nom)
            score_logging[i] = base_score
            U_traj_logging[i] = U_traj_nom

            #put some convergence criteria in here later

            
        return U_traj_nom

            

        



        
    def sort_A(self):
        sorting_mat = np.ones((self.state_dim)) - self.sort_vec
        self.A_sorted = self.A_lin * sorting_mat
        

 
    def get_sort_vec(self):
        for mat in range(len(self.picks)):
            for idx, i in enumerate(self.picks[mat]):
                self.sort_vec[i*self.num_materials + mat] = 1
            if self.sort_vec.T @ self.state_estimate > self.picks_per_sec / self.update_rate:
                self.sort_vec *= (self.picks_per_sec/self.update_rate)/(self.sort_vec.T @ self.state_estimate)
        
        # self.sort_vec[(self.num_volumes-1)*self.num_materials : self.num_volumes*self.num_materials] = 1 ## negative sort going off the end of the line
        
        # def get_sort_vec(self):
    #     for mat in range(len(self.picks)):
    #         for idx, i in enumerate(self.picks[mat]):
    #             if self.state_estimate[i*self.num_materials + mat] < self.picks_per_step_per_volume_max: #is there less material at a location than the max pick rate
    #                 self.sort_vec[i*self.num_materials + mat] = 1
    #             else: 
    #                 self.sort_vec[i*self.num_materials + mat] = self.picks_per_step_per_volume_max / self.state_estimate[i*self.num_materials + mat]
        self.cum_sorts += self.state_estimate.T @ self.sort_vec       
        # print(self.sort_vec)

    
    def speed_msg_callback(self, speed_msg):
        
        
        self.speed_request_block = 0.01301793014 * speed_msg + 0.2946163136
        
        

E = estimator()
```

```python
sort_prim = np.zeros((E.state_dim))
pics_per_step = E.picks_per_sec / E.update_rate
state_dim_ones = np.ones((E.state_dim))
# score_logging = []
for mat in range(len(E.picks)):
    for idx, i in enumerate(E.picks[mat]):
            sort_prim[i*E.num_materials + mat] = 1
        
def rollout_linearize_A(x):

    # print('x in rollout linearizer: ', x)
    floor = np.floor(x[-1])
    # print(floor)
    ceil = floor + 1
    ratio = x[-1] - floor
    A = E.A[:, :, int(floor)] * (1-ratio) + E.A[:, :, int(ceil)] * ratio
    
    for mat in range(len(E.picks)):
        # for idx, i in enumerate(E.picks[mat]):
        #     sort_vec[i*E.num_materials + mat] = 1
        ideal_sorts = sort_prim.T @ x
        if ideal_sorts > pics_per_step:
            
            A = A * (state_dim_ones - (pics_per_step/ideal_sorts * sort_prim))
        else:
            A = A * (state_dim_ones - sort_prim)
    
    return A, sort_prim

def rollout(x, U_traj):
    # print('u traj beginning of rollout: ', U_traj)
    # print(x)
    X_traj = np.zeros((E.horizon+1, E.state_dim))
    score = 0
    for step in range(E.horizon):
        
        if step == 0:
            # print('step: ', step, 'x_init: ', x)
            A_rollout, sort_vec = rollout_linearize_A(x)
            # print(A_rollout, x, E.B, U_traj[step])
            X_traj[step] = A_rollout @ x + E.B * U_traj[step]
            # print('after doing stuff: ', X_traj[step])
        else:
            # print('step: ', step, 'x_traj: ', X_traj[step-1])
            A_rollout, sort_vec = rollout_linearize_A(X_traj[step-1])
            X_traj[step] = A_rollout @ X_traj[step - 1] + E.B * U_traj[step]

        
        score_vec = E.prices[0] * sort_vec
        score_vec[(E.num_volumes-1)*E.num_materials : E.num_volumes*E.num_materials] = E.prices[1]
        score_vec[(E.num_volumes-1)*E.num_materials] -= E.prices[0]

        score += score_vec @ X_traj[step]
        # print('score iun rollour: ', score)
        # print('control this step: ', U_traj[step])
        # print('belt speed: ', X_traj[step, -1])
    A_rollout, sort_vec = E.rollout_linearize_A(X_traj[E.horizon-1])
    X_traj[E.horizon] = A_rollout @ X_traj[E.horizon-1] + E.B * U_traj[E.horizon-1]
    score_vec = E.prices[0] * sort_vec
    score_vec[(E.num_volumes-1)*E.num_materials : E.num_volumes*E.num_materials] = E.prices[1]
    score_vec[(E.num_volumes-1)*E.num_materials] -= E.prices[0]
    score += score_vec @ X_traj[E.horizon]
    
    return score, X_traj
    
def solve_opt_traj(x_init):

    #roll out nom traj, nom traj + delta
    U_traj_nom = np.ones((E.horizon)) * x_init[-1]
    grad = np.zeros_like(U_traj_nom)
    deltas = E.delta * np.diag(np.ones_like(U_traj_nom))
    lr = E.learning_rate
    score_logging = np.zeros((E.iters))
    U_traj_logging = np.zeros((E.iters, E.horizon))
    # print('deltas: ', deltas)
    

    base_score, x_t = rollout(x_init, U_traj_nom)
    # print('base score: ', base_score) 
    for i in range(E.iters):
        #get gradient about u_traj_nom
        
        for dir in range(E.horizon):
            s, x_t = rollout(x_init, U_traj_nom + deltas[dir])
            # print('s: ', s)
            grad[dir] = (s-base_score)/E.delta
            
        # print(grad)
        U_traj_nom += lr * grad
        U_traj_nom.clip(0, 4)
        base_score, x_t = rollout(x_init, U_traj_nom)
        score_logging[i] = base_score
        U_traj_logging[i] = U_traj_nom

        #put some convergence criteria in here later

        
    return U_traj_nom

def rollout_score(U_traj_nom):
    score, x = rollout(x_init, U_traj_nom)
    return 1/score

def solve_opt_traj_scipy(x_init):
    U_traj_nom = np.ones((E.horizon)) * x_init[-1]
    bnds = [(.6, 1.3) for i in range(E.horizon)]
    # res = scipy.optimize.minimize(rollout_score, U_traj_nom, bounds = bnds, method='Nelder-Mead')
    res = scipy.optimize.minimize(rollout_score, U_traj_nom)
    return res
```
